# GEO Prompt Audit Analysis

Minimal notebook for summarizing prompt-level AI visibility results.

```python
import pandas as pd

df = pd.read_csv('data/sample-results.csv')
df.head()
```

```python
overall_summary = pd.DataFrame({
    'metric': ['Average score', 'Brand mention rate (%)'],
    'value': [df['score'].mean(), df['brand_mentioned'].eq('yes').mean() * 100],
})
overall_summary.round(1)
```

```python
by_engine = (
    df.assign(brand_mentioned_yes=df['brand_mentioned'].eq('yes'))
      .groupby('engine', as_index=False)
      .agg(average_score=('score', 'mean'), mention_rate=('brand_mentioned_yes', 'mean'))
)
by_engine['mention_rate'] = by_engine['mention_rate'] * 100
by_engine.round(1).sort_values('average_score', ascending=False)
```

```python
by_category = (
    df.assign(brand_mentioned_yes=df['brand_mentioned'].eq('yes'))
      .groupby('prompt_category', as_index=False)
      .agg(average_score=('score', 'mean'), mention_rate=('brand_mentioned_yes', 'mean'))
)
by_category['mention_rate'] = by_category['mention_rate'] * 100
by_category.round(1).sort_values('average_score', ascending=False)
```

```python
competitor_mentions = (
    df.loc[df['competitors_mentioned'].ne('none'), 'competitors_mentioned']
      .str.split(',')
      .explode()
      .str.strip()
      .value_counts()
      .rename_axis('competitor')
      .reset_index(name='mentions')
)
competitor_mentions
```

Use these summary tables to compare engines and prompt categories, then review competitors that appear when the target brand is missing.
